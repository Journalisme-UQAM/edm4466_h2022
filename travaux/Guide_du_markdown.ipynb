# Guide de référence du _markdown_

## Qu'est-ce que le markdown?

Le markdown est ce qu'on appelle un « préprocesseur » HTML. Il permet de formater du texte en ligne plus facilement qu'en utilisant du HTML.

Ce guide de référence est une adaptation en français du [guide fourni par Google](https://colab.research.google.com/notebooks/markdown_guide.ipynb).

<hr>

## Mise en forme générale

Markdown | Aperçu
--- | ---
`**gras**` ou `__gras__` | **gras** 
`*italique*` ou `_italique_` | *italique*
`**_gras italique_**` ou `*__gras italique__*` | _**gras italique**_
`~~barré~~` | ~~barré~~

<br>

---

## Liens

Pour créer un lien en markdown, il suffit de placer le texte qui sera cliquable entre crochets `[ ]`, qu'on fait immédiatement suivre de l'URL vers lequel ce lien nous mènera entre parenthèses `( )`.

Voici un exemple.

Ce texte en markdown:

```markdown
Le [programme de journalisme de l'UQAM](https://journalisme.uqam.ca) est en phase avec son époque.
```

Produira le rendu suivant:

Le [programme de journalisme de l'UQAM](https://journalisme.uqam.ca) est en phase avec son époque.


## Images

Pour insérer des images dans vos cellules de texte, la syntaxe à utiliser en markdown ressemble beaucoup à celle des liens. À une différence près.

Il s'agit de commencer par un point d'exclamation. Le texte entre les crochets sera celui qui sera affiché si un problème survient. L'URL vers l'image est placé entre parenthèses.

Voici un exemple. Le texte suivant en markdown:

```
![Logo du programme de journalisme de l'UQAM](https://journalisme.uqam.ca/wp-content/uploads/sites/77/2022/01/JournalismeUQAM-logo-100px.png)```

Sera rendu ainsi:

![Logo du programme de journalisme de l'UQAM](https://journalisme.uqam.ca/wp-content/uploads/sites/77/2022/01/JournalismeUQAM-logo-100px.png)

## Entêtes

<!-- Vous pouvez utiliser du HTML pur, aussi, si le cœur vous en dit -->

<table>

<thead>
<th>Markdown</th>
<th>Aperçu</th>
</thead>

<tr>
<td style="text-align:left;"><code>
# Niveau 1
</code></td>
<td>
<h1>Niveau 1</h1>
</td>
</tr>

<tr>
<td><code>
## Niveau 2
</code></td>
<td>
<h2>Niveau 2</h2>
</td>
</tr>

<tr>
<td><code>
### Niveau 3
</code></td>
<td>
<h3>Niveau 3</h3>
</td>
</tr>

</table>

Comme en HTML, vous pouvez utiliser jusqu'à six niveaux d'entêtes.

L'utilisation de ces entêtes génère une table des matières (ou sommaire) que vous pouvez voir à gauche de votre carnet, un peu comme ce que les styles permettent dans Word ou LibreOffice ou ce que les titres génèrent dans Google Docs.

Voici les mêmes niveaux en markdown (double-cliquez pour voir le code qui sous-tend la cellule):

# Niveau 1

## Niveau 2

### Niveau 3

---

## Listes

### Listes non-ordonnées

<table>

<thead>
<th>Markdown</th>
<th>Aperçu</th>
</thead>

<tr>
<td><code>
* Et de un <br>
* Et de deux <br>
* Et de trois
</code></td>
<td>
<ul>
<li>Et de un</li>
<li>Et de deux</li>
<li>Et de trois</li>
</ul>
</td>
</tr>

</table>
<br>

Rendu hors tableau (double-cliquez pour voir le code markdown qui sous-tend la cellule):

* Et de un
* Et de deux
* Et de trois 
<br>

---

### Listes ordonnées

<table>

<thead>
<th>Markdown</th>
<th>Aperçu</th>
</thead>

<tr>
<td style="text-align:left;"><code>
1. Un <br>
1. Un (en fait, deux) <br>
1. Un (en fait, trois)
</code></td>
<td>
<ol>
<li>Un</li>
<li>Un (en fait, deux)</li>
<li>Un (en fait, trois)</li>
</ol>
</td>
</tr>

</table>
<br>

Rendu hors tableau :

1. Un
1. Un (en fait, deux)
1. Un (en fait, trois)
<br>

---

### Liste avec niveaux (indentations)

Ici, il est plus clair de faire la démonstration sans l'insérer dans un tableau.

```markdown
* Premier item, qui en comprend deux autres:
  * Deux espaces avant l'astérisque suffisent
  * Et ça c'est le 2e sous-item, qui en comprend trois derniers:
    1. Uno
    1. Dos
    1. Tres
```

## Tableaux

Markdown permet de faire assez rapidement des tableaux simples.

```markdown
Colonne 1  | Colonne 2 | Colonne 3 
-----|-----|-----
Ligne 1 col 1 | Ligne 1 col 2 | Ligne 1 col 3
Ligne 2 col 1 | Ligne 2 col 2 | Ligne 2 col 3
Ligne 3 col 1 | Ligne 3 col 2 | Ligne 3 col 3
```

## Code

Markdown vous premet aussi d'inscrire du code dans une cellule de texte.

### Dans une phrase

Vous pouvez insérer un peu de code dans une phrase en le plaçant entre deux accents graves. C'est pratique pour parler de variables (avec des noms courts comme `x` ou longs comme `supercalifragilistique`), de fonctions (comme `print()`), ou autres.

### Blocs de code

Pour écrire un bloc de code plus long, il faut l'insérer entre trois accents graves au début, comme ceci:

<code> ``` </code>

Et trois autres à la fin, comme cela:

<code> ``` </code>

Vous pouvez même accompagner les trois premiers accents graves d'un nom de langage pour colorer le code en fonction du langage qui vous intéresse. Voici quelques exemples:

##### **Python**

```python
for bidule in bidules:
  if bidule[0] == "Saperlipopette":
    print(bidule[-1])
```

##### **JavaScript**

```javascript
// Ceci est un commentaire
console.log("Bonjour/Hi");
```

##### **HTML**

```html
<table>
  <thead>
    <th>Village</th>
    <th>Gentilé</th>
  </thead>
  <tr>
    <td style="text-align:left;">
      Mayo
    </td>
    <td>
      Mayolois.es
    </td>
  </tr>
</table>
```

##### **CSS**

```css
.auteur {
  font-family: Comic Sans;
  padding: 10px;
}
```
